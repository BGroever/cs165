```python
import re
import matplotlib.pyplot as plt

START_INDEX=0
END_INDEX=100
time = []
for index in range(START_INDEX,END_INDEX):
    filename = '/home/ben/cs165/src/benchmark/output/server_'+str(index).zfill(2)+'.out'
    textfile = open(filename, 'r')
    filetext = textfile.read()
    textfile.close()
    matches = re.findall("(\d+\.\d+)", filetext)
    time.append(float(matches[-2]))
    
data_length = np.linspace(0,1000000,101,dtype=np.int)[1:]
plt.plot(data_length,time)
plt.xlabel('Number of rows')
plt.ylabel('Load Time (ms)')
plt.rc('font', size=13)
plt.savefig('benchmark3.png')
```

```python
START_INDEX=200
END_INDEX=400
time_primary = []
time_secondary = []
for index in range(START_INDEX,END_INDEX):
    filename = '/home/ben/cs165/src/benchmark/output3/server_'+str(index).zfill(2)+'.out'
    textfile = open(filename, 'r')
    filetext = textfile.read()
    textfile.close()
    matches = re.findall("(\d+\.\d+)", filetext)
    time_primary.append(float(matches[-2]))
    time_secondary.append(float(matches[-1]))

data_length = np.linspace(0,200000,201,dtype=np.int)[1:]/1000000
plt.plot(data_length,time_primary)
plt.plot(data_length,time_secondary)
plt.legend(['primary','secondary'])
plt.xlabel('Qualifing rows/Total rows')
plt.ylabel('Time (ms)')
plt.rc('font', size=13)
plt.savefig('benchmark5.png')
```
