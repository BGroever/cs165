```python
import matplotlib
%matplotlib inline
import re
import numpy as np
import matplotlib.pyplot as plt


START_INDEX=0
END_INDEX=100
intructions = []
for index in range(START_INDEX,END_INDEX):
    filename = './benchmark/output1_new/server_'+str(2000+index)+'.err'
    textfile = open(filename, 'r')
    filetext = textfile.read()
    textfile.close()
    matches = re.findall("(\d+,?\d+,?\d+)\s+cy", filetext)
    intructions.append(int(matches[0].replace(',', '')))
    #print(matches[0].replace(',', ''))

data_length = np.linspace(0,1000000,101,dtype=np.int)[1:]
plt.plot(data_length,intructions)
plt.xlabel('Number of rows')
plt.ylabel('Cycles')
#plt.ylim([0,5.5])
plt.rc('font', size=13)
plt.savefig('benchmark1_cycles.png')
```
